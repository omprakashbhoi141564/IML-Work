```python
import numpy as np
```

```python
x1=np.array([3,1,2,4,3])
x2=np.array([6,3,4,7,5])
y=np.array([1,0,0,1,1])
```

```python
X=np.column_stack((np.ones(5),x1,x2))
```

```python
B=np.linalg.inv(X.T@X)@X.T@y
```

```python
b0,b1,b2=B
```

```python
print("the constant value of (b0) is:",b0)
print("the constant value of (b1) is:",b1)
print("the constant value of (b2) is:",b2)
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
z = X @ B
print("\nLinear combination (z):")
print(z)
```

```python
y_prob = np.array([sigmoid(val) for val in z])
print("\nPredicted Probabilities (Sigmoid_value):")
print(y_prob)
```

```python
y_pred = (y_prob > 0.5).astype(int)
print("\nPredicted Class Labels:")
print(y_pred)
```

```python
TP = np.sum((y_pred == 1) & (y == 1))
TN = np.sum((y_pred == 0) & (y == 0))
FP = np.sum((y_pred == 1) & (y == 0))
FN = np.sum((y_pred == 0) & (y == 1))
```

```python
print("\nConfusion Matrix:")
print(f"TP = {TP}, FP = {FP}")
print(f"FN = {FN}, TN = {TN}")
```

```python
# Accuracy
Accuracy = (TP + TN) / (TP + TN + FP + FN)
# Precision
Precision = TP / (TP + FP)
# Recall (Sensitivity)
Recall = TP / (TP + FN)
# F1_Score
F1_Score = 2 * (Precision * Recall) / (Precision + Recall)
```

```python
print("\nPerformance Metrics:")
print(f"Accuracy = {Accuracy}")
print(f"Precision = {Precision}")
print(f"Recall (Sensitivity) = {Recall}")
print(f"F1_Score = {F1_Score}")
```

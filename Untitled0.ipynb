```python
import numpy as np
```

```python
x1=np.array([1,2,3,4,5,6])
x2=np.array([60.0,64.8,84.0,59.4,108.0,75.0])
y=np.array([18.4,21.6,17.6,16.0,17.6,19.6])
```

```python
X=np.column_stack((np.ones(6),x1,x2))
```

```python
B=np.linalg.inv(X.T@X)@X.T@y
```

```python
b0,b1,b2=B
```

```python
print("the constant value of (b0) is:",b0)
print("the constant value of (b1) is:",b1)
print("the constant value of (b2) is:",b2)
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
z = X @ B
print("\nLinear combination (z):")
print(z)
```

```python
y_prob = np.array([sigmoid(val) for val in z])
print("\nPredicted Probabilities (Sigmoid_value):")
print(y_prob)
```

```python
import numpy as np
```

```python
X1=np.array([1,2,1])
X2=np.array([2,2,1])
Y=np.array([6,9,4])
```

```python
X=np.column_stack((np.ones(3),X1,X2))
```

```python
B=np.linalg.inv(X.T @ X) @ X.T @ Y
```

```python
b0,b1,b2= B
```

```python
print("Intersept(bo)",b0)
print("Coefficient for Studied hours(b1)",b1)
print("Coefficient for Sleep hours(b2)",b2)
```

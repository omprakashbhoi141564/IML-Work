```python
import numpy as np
```

```python
x=np.array([1,2])
y=np.array([1,3])
```

```python
X=np.column_stack((np.ones(2),x))
```

```python
B=np.linalg.inv(X.T@X)@X.T@y
```

```python
b0,b1=B
```

```python
print("the constant value of (b0) is:",b0)
print("the constant value of (b1) is:",b1)
```

```python
y_prob = b0+b1*x
```

```python
for lam in[0,1,5,100]:
  I = np.eye(X.shape[1])
  I[0,0] = 0
  B_ = np.linalg.inv(X.T@X+lam*I)@X.T@y
  print(f"lambda = {lam},Coefficient = {B}")
```

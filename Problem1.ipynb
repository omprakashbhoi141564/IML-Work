```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
# Provided Data
X = [1,2,3,4,5]
Y = [2,4,5,4,5]
```

```python
# Create a dictionary where keys are column name and value are the list
data={'X':X, 'Y':Y}
```

```python
# create pandas data frame
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
%matplotlib inline
plt.scatter(df.X,df.Y)
plt
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['X']],df.Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```
